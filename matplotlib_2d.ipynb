# Matplotlib 

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
randomnumber = np.arange(10)
randomnumber
```

```python
random_number = np.random.rand(10)
random_number
```

```python
plt.plot(random_number, label='Line one', color='g')
plt.xlabel('Range')
plt.ylabel('Numbers')
plt.title('First Plot')
```

```python
style.use('ggplot') # style gives the grid representation for the plot
plt.plot(random_number, label='Line one', color='r', linewidth=2)
plt.xlabel('Range')
plt.ylabel('Numbers')
plt.title('First Plot')
# lable='line one' is not reflected without legend, as this is legend label
```

```python
style.use('ggplot') # style gives the grid representation for the plot
plt.plot(random_number, label='Line one', color='b', linewidth=2)
plt.xlabel('Range')
plt.ylabel('Numbers')
plt.title('First Plot')
plt.show()
```

# Pie chart for ohio state- leading causes for death

```python
# import libraries
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load the data
# we need to two list one for causes of death and another for percentile for each causes
# causes
causes = 'Chronic diseases', 'Unintentional injuries', 'Alzheimers', 'Infuenza and Pneumonia', 'Sepsis', 'Others'
# Percentile of death for each causes
percentile = [62, 5, 4, 2, 1, 26]
```

```python
#set the pie chart plot and properties
# set the figure size
plt.figure(figsize=(6,6))
# explode the larget pie in the data set 
explode = (0.05,0,0,0,0,0)
# set the pie chart properties
plt.pie(percentile, labels=causes, explode=explode,autopct='%1.1f%%', startangle=50)
# set axis, to draw a pie chart as a circle
plt.axis('equal')
#set the title for the pie chart
plt.title('Ohio state - 2015: Leading Causes for death')
#show the pie chart
plt.show()
```

# Various plots using matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('./Datasets_4754/OSL Datasets/Lesson_03/HousePrices.csv',parse_dates=['date'], index_col=['date'])
df.sort_index(inplace=True)
```

```python
df.head()
```

## Line chart

```python
plt.plot(df['price'])
```

## Boxplot

```python
plt.boxplot(df['price'])
```

## Histogram

```python
plt.hist(df['sqft_living'])
```

## Scatter plot

```python
plt.xlabel('Price')
plt.ylabel('Sqft Living')
plt.title('Price vs Sqft living')
plt.scatter(x=df['price'], y=df['sqft_living'])
```

# Bar and pie chart from a dict

```python
dict_a = {'a': 10, 'b': 15, 'c': 30}
```

```python
plt.bar(x=dict_a.keys(), height=dict_a.values())
```

```python
plt.pie(x=dict_a.values(), labels=dict_a.keys())
```
