# Seaborn 3D

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../Datasets_4754/OSL Datasets/Lesson_03/ADANIPORTS.csv',parse_dates=True)
df.head()
```

```python
# calculate two columns
df['H-L'] = df.High - df.Low
df['100MA'] = df['Close'].rolling(100).mean()  # moving average
```

## Scatter plot with dark grid style

```python
sns.set_style('darkgrid')
```

```python
#3d plot properties 
ax = plt.axes(projection='3d')
ax.scatter(df.index, df['H-L'], df['100MA'])
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('100MA')
plt.show()
```

## plot3D with white grid style

```python
import numpy as np
```

```python
z1 = np.linspace(0,10,100)
x1 = np.cos(2*z1)
y1 = np.sin(2*z1)
```

```python
sns.set_style('whitegrid')
ax = plt.axes(projection='3d')
ax.plot3D(x1,y1,z1)
plt.show()
```

```python
##spiral can be increased by increasing x1 and y1 axis
z1 = np.linspace(0,10,100)
x1 = np.cos(4*z1)
y1 = np.sin(4*z1)

sns.set_style('whitegrid')
ax = plt.axes(projection='3d')
ax.plot3D(x1,y1,z1)
plt.show()
```

## Plot surface with mesh grid and whitegrid style

```python
x1,y1 = np.linspace(-5,5,50), np.linspace(-5,5,50)
```

```python
def return_z(x,y):
    return 50-(x**2+y**2)
```

```python
z1 = return_z(x1,y1)
z1
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(x1,y1,z1)
plt.show()
### gives error because fro plot_surface, z1 should be two dimensional
```

```python
x1,y1 = np.meshgrid(x1,y1) ## to convert z1 to two dimensional
z2 = return_z(x1,y1)
z2
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(x1,y1,z2)
plt.show()
```
