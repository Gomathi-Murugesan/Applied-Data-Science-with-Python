# Seaborn 

```python
import seaborn as sns
```

```python
sns.set_theme()
```

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.relplot(data=fmri,kind="line", x="timepoint", y="signal", col='region', hue="event", style="event")
```

## Analysis

- Two graphs based on region, parietal, frontal
- Each graph has two plots, one for stim, and one for cue --- events
- X and y axis are same for both the graphs
- Signals are high for parietal than frontal
- Both the plot peaks at timepoint 5.0
- seaborn does statistical calculations and gave the error bar
- error bar is wider during the peaks
