```python
import pandas as pd
import nuampy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

sns.set_style('darkgrid')
```

```python
df_drugsales = df[['Drug1_Sales','Drug2_Sales','Drug3_Sales','Drug4_Sales','Drug5_Sales','Drug6_Sales','Drug7_Sales','Drug8_Sales']]
df_drugsales
```

```python
import statistics
def avg_sales(df):
    avg_sales = []
    sales_list = []
    for keys,values in df.iteritems():
        #print(keys)
        #print(statistics.mean(values))
        sales_list.append(keys)
        avg_sales.append(float(statistics.mean(values)))
    return sales_list, avg_sales
sales_list, avg_sales = avg_sales(df_drugsales)
print(sales_list, avg_sales)
```

## Monthly Average Sales of all drugs

```python
plt.figure(figsize=(5,5))
explode = (0,0,0,0.05,0,0,0,0)
plt.pie(avg_sales, labels=sales_list, explode=explode, autopct='%1.1f%%', startangle=50)
plt.axis='equal' # to draw pie chart as a circle
plt.title('Drug Sales of September 2019')
plt.show()
```

- The pie chart is draw for the average sales of the drugs.
- The chart shows that drug4 has maximum sales of 41.6% and drug7 has the least sales of 3.8%

```python
for keys,values in df_drugsales.iteritems():
    plt.tick_params(axis='x', rotation=45)
    plt.plot(df.index, values)
    plt.xlabel('Dates')
    plt.ylabel(keys)
    plt.title(f'Plot of Dates vs {keys}')
    plt.show()
```

# Weekly Analysis

```python
wdf = df.resample('W').mean()
```

```python
wdf.head()
```

```python
wdf.tail()
```

```python
wdf.shape
```

- weekly dataframe has 6 rows, 6 weeks totally
- mean of each and every column for that week is calculated

```python
wdf.describe()
```

### Inference to weekly analysis
- Drug4 tops the sales
- Drug6 sold the least units
- Drug4 sales is wrost of all even drug6 unit sales is 1

```python
plt.figure(figsize=(6,3), dpi=100)
plt.tick_params(axis='x', rotation=45)
plt.title('Drug1 weekly report')
plt.bar(x=wdf.index, height=wdf['Drug1_Sales'])
```

### Inference of Drug1 weekly sales:
- Best week was from 09/17 to 09/21
- Sales peaked at 09/21 and had some decline


```python
for drug in drugs:
    plt.figure(figsize=(6,3), dpi=100)
    plt.tick_params(axis='x', rotation=45)
    plt.title(f'{drug}_Sales weekly report')
    plt.bar(x=wdf.index, height=wdf[drug+'_Sales'])
```
